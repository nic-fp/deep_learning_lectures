```python
import torch
```

```python
class myModel(torch.nn.Module):
    #use a param that is the list of each layer size
    def __init__(self, layer_size = [512,512,512]):
        super(myModel,self).__init__()
        layers = []
        #flatten layer will take any image an flatten to vector!
        c = 128*128*3 #channels, initial is our size
        layers.append(torch.nn.Flatten())#?
        for i in layer_size:
            #initially go 128*128*3 to 512 then 512 to 512 etc
            layers.append(torch.nn.Linear(c, i))
            layers.append(torch.nn.ReLU())
            c = i
        #last layer needs to map to output size
        #no relu!
        layers.append(torch.nn.Linear(c,102))
        #sequential will create a network where layers are called in order
        self.model = torch.nn.Sequential(*layers)
    def forward(self, x):
        #forward is us just calling the model
        return self.model(x)

```

```python
#as we add layers output gets increasingly smaller

#at a certain ponit they stop getting smaller bc of bias term
# to see selt bias = False in Linear layers
for n in range(10):
    netn = myModel([512]*n)
    x = torch.randn(10,3,128,128)
    print(f"{n=} {netn(x).norm()=}")
```

```python
class myModelNormalization(torch.nn.Module):
    #use a param that is the list of each layer size
    def __init__(self, layer_size = [512,512,512]):
        super(myModelNormalization,self).__init__()
        layers = []
        #flatten layer will take any image an flatten to vector!
        c = 128*128*3 #channels, initial is our size
        layers.append(torch.nn.Flatten())#?
        for i in layer_size:
            #initially go 128*128*3 to 512 then 512 to 512 etc
            layers.append(torch.nn.Linear(c, i))
            layers.append(torch.nn.BatchNorm1d(i))
            layers.append(torch.nn.ReLU())
            c = i
        #last layer needs to map to output size
        #no relu!
        layers.append(torch.nn.Linear(c,102))
        #sequential will create a network where layers are called in order
        self.model = torch.nn.Sequential(*layers)
    def forward(self, x):
        #forward is us just calling the model
        return self.model(x)

```

```python
#as we add layers output gets increasingly smaller

#at a certain ponit they stop getting smaller bc of bias term
# to see selt bias = False in Linear layers
#now with normalization activations stay high
for n in range(10):
    netn = myModelNormalization([512]*n)
    x = torch.randn(10,3,128,128)
    print(f"{n=} {netn(x).norm()=}")
```

```python
class myModelLayerNormalization(torch.nn.Module):
    #use a param that is the list of each layer size
    def __init__(self, layer_size = [512,512,512]):
        super(myModelLayerNormalization,self).__init__()
        layers = []
        #flatten layer will take any image an flatten to vector!
        c = 128*128*3 #channels, initial is our size
        layers.append(torch.nn.Flatten())#?
        for i in layer_size:
            #initially go 128*128*3 to 512 then 512 to 512 etc
            layers.append(torch.nn.Linear(c, i))
            layers.append(torch.nn.LayerNorm(i))
            layers.append(torch.nn.ReLU())
            c = i
        #last layer needs to map to output size
        #no relu!
        layers.append(torch.nn.Linear(c,102))
        #sequential will create a network where layers are called in order
        self.model = torch.nn.Sequential(*layers)
    def forward(self, x):
        #forward is us just calling the model
        return self.model(x)

for n in range(50):
    netn = myModelLayerNormalization([512]*n)
    x = torch.randn(10,3,128,128)
    print(f"{n=} {netn(x).norm()=}")
```

```python
class myModelLN(torch.nn.Module):
    #need this bc we are donig skip connections so no longer sequential
    class Block(torch.nn.Module):
        def __init__(self, in_channels, out_channels):
            super().__init__()
            self.lin = torch.nn.Linear(in_channels, out_channels)
            self.norm = torch.nn.LayerNorm(out_channels)
            self.relu = torch.nn.ReLU()
            #this is what we talked about in lectures
            #don't want to do much else defeats purpose
            #just a resize to the skip
            if in_channels != out_channels:
                self.skip = torch.nn.Linear(in_channels, out_channels)
            else:
                #do nothing, just add directly 
                self.skip = torch.nn.Identity()


        def forward(self, x): 
            # can't do return self.relu(self.norm(self.lin(x))) + x
            #bc in the first layer it gets flattened and then can't add original tensor
            #for first layer can't have skip connection
            #first block in linear layer took the 128*128*3 input and made it 512
            
            #easy fix -> only add residual when conveinient
            #not common fix
            #instead when we create blcok
            #check if input and output channels are same
            y = self.relu(self.norm(self.lin(x)))
            # if y.shape == x.shape:
            #     return x + y

            # return  y
            #either does nothing or reshapes then adds
            return self.skip(x) + y

    #use a param that is the list of each layer size
    def __init__(self, layer_size = [512,512,512]):
        super().__init__()
        layers = []
        #flatten layer will take any image an flatten to vector!
        c = 128*128*3 #channels, initial is our size
        layers.append(torch.nn.Flatten())#?
        #first layer is outside of blcok
        #typically conv or embedding
        layers.append(torch.nn.Linear(c, 512, bias=False))
        c = 512
        for i in layer_size:
            #initially go 128*128*3 to 512 then 512 to 512 etc
            layers.append(self.Block(c, i))
            c = i
        #last layer needs to map to output size
        #no relu!
        layers.append(torch.nn.Linear(c,102))
        #sequential will create a network where layers are called in order
        self.model = torch.nn.Sequential(*layers)
    def forward(self, x):
        #forward is us just calling the model
        return self.model(x)

for n in range(10):
    netn = myModelLN([512]*n)
    x = torch.randn(10,3,128,128)
    print(f"{n=} {netn(x).norm()=}")
```
