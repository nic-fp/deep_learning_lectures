```python
a = torch.arange(6).view(3,2,1,1)
#ways to remove last dim
print(a[...,0].shape)
#squeeze removes dims that are size one  so this works
print(a.squeeze(-1).shape)
#but this does nothing
print(a.squeeze(0).shape)

#NEVER do this, but this gets rid of all dims of size one
print(a.squeeze().shape)
```

```python
a = torch.arange(4)
b = torch.arange(5)

try:
    print(a + b)
except:
    print('This fails')

#if dim is 1 then it works. Just adds scalar everywhere
a = torch.arange(1)
b = torch.arange(5)

try:
    print(a + b)
except:
    print('This fails')

#why does this work?
#broadcasting
#duplicates a along 2nd dim (the size 1) and b along 1st
a = torch.rand((4,1))
b = torch.rand((1,5))

try:
    print(a + b)
    print((a+b).shape)
except:
    print('This fails')
```

```python
#how to do elementwise distance of all elements without broadcasting??

x = torch.randn(10,2)
d = torch.zeros(10,10)

for i in range(10):
    for j in range(10):
        d[i,j] = (x[i] - x[j]).pow(2).sum()

print(d.sum())

#with broadcasting
#add singleton in 2nd dim and subtract by one with singleton in 1st. 
#subtracts any row in x from any other row
# so it gives d[i,j] = x[i] - x[j] for every i, j
d = (x[:, None, :] - x[None,:,:]).pow(2).sum()
print(d)
```

```python
#what about max distance? 
d = (x[:, None, :] - x[None,:,:]).pow(2)
print(d.max(), (d.argmax()//d.shape[0], d.argmax()%d.shape[0]))

#argmax flattens first, so to get point use floor to get row and mod to get column
```
