```python
import torch
```

```python
model = torch.nn.Linear(10,1)
```

```python
x = torch.randn(20,10)
y = torch.randn(20, 1)
print(f'{ x[0] =}')
print(f'{y[0] = }')
```

```python
pred_y = model(x)
print(f'{pred_y[0] = }')
```

```python
loss = torch.nn.functional.mse_loss(pred_y, y)
print(f"{loss = }")
```

```python
def mse_loss(y_pred, y):
    return torch.mean((y_pred-y)**2)

print(mse_loss(pred_y, y))
```

```python
loss = torch.nn.functional.binary_cross_entropy_with_logits(pred_y, y)
#always use with logits as it will allow us to use linear model without addl transformations bc it will wrap in sigmoid for us
print(f"{loss = }")
#.7 means 50/50 guess so this is worse than guessing
```
