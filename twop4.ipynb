```python
import matplotlib.pyplot as plt
import torch


def f(x: torch.Tensor()) -> torch.Tensor:
    return 3*x**2+torch.sin(5*x)

def grad_f(x:torch.Tensor)->torch.Tensor:
    return 6*x+5*torch.cos(5*x)


plt.plot(torch.arange(-1,1,.01), f(torch.arange(-1,1,.01)))
```

```python
x = torch.randn(1)
print(x)
all_x = []
for _ in range(10000):
    x = x - .01 *grad_f(x)
    all_x.append(x)

plt.plot(torch.arange(-1,1,.01), f(torch.arange(-1,1,.01)))
plt.plot(all_x, f(torch.stack(all_x)), 'ro')

print(x)
#depending on init will find either global or the local min
```
