```python
import torch
```

```python
x = torch.rand(10, requires_grad =True)
y = torch.rand(10)
print(x)
print(y)
```

```python
#has a grad fn attached
b=(x**2).mean()
print(b)
#populates x's gradient
b.backward()
```

```python
print(x/10 * 2)
print(x.grad)
```

```python
try:
    b.backward()
except:
    print("You can't compute backward again")
    print('need to redefine b')
    print("IMPORTANT this will sum up the gradient of x!!")

#has a grad fn attached
b=(x**2).mean()
#populates x's gradient
b.backward()
print(x/10 * 2)
print(x.grad)
print('x.grad changed')
```

```python
x = torch.randn(1000000, requires_grad = True, device = 'cuda')
torch.cuda.memory_allocated() /1024/1024 #num MB
```

```python
for _ in range(10):
    b = torch.relu(x)
print(torch.cuda.memory_allocated() /1024/1024) #num MB


b.sum().backward()

torch.cuda.memory_allocated() /1024/1024
```
